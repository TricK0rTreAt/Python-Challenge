```python
import os
import pandas as pd
import csv

#----------------------------PyBank----------------------------
df=pd.read_csv("budget_data.csv")

#export a text file with the results

path = 'Financial Analysis.txt'
f = open(path, 'w')
print("Financial Analysis")
print("Financial Analysis", file=f)

print("----------------------------")
print("----------------------------", file=f)

#The total number of months included in the dataset
total_months=df["Date"].count()
print("total months = ",total_months)
print("total months = ",total_months, file=f)

#The net total amount of "Profit/Losses" over the entire period
total_amount=df["Profit/Losses"].sum()
print("total amount = $",total_amount)
print("total amount = $",total_amount, file=f)

#add a new column
df["Amount_changed"]=0

#The changes in "Profit/Losses" over the entire period, and then the average of those changes
i=0
Beginning_amount=0
Final_amount=0

for i in range(0,85):
    Beginning_amount=df.iloc[i,1]
    Final_amount=df.iloc[i+1,1]
    df.iloc[i+1,2]=Final_amount - Beginning_amount
    df.iloc[0,2]=0
    
Average_changes=round(df["Amount_changed"].sum()/(total_months-1),2)
print("Average changes = $",Average_changes)
print("Average changes = $",Average_changes, file=f)

# The greatest increase in profits (date and amount) over the entire period

j=0

for j in range(0,85):
    if df.iloc[j,2]==df["Amount_changed"].max():
        print("The greatest increase in profits is $", df["Amount_changed"].max(), "on", df.iloc[j,0],".")
        print("The greatest increase in profits is $", df["Amount_changed"].max(), "on", df.iloc[j,0],".", file=f)
        
# The greatest decrease in profits (date and amount) over the entire period
k=0

for k in range(0,85):
    if df.iloc[k,2]==df["Amount_changed"].min():
        print("The greatest decrease in profits is $", df["Amount_changed"].min(), "on", df.iloc[k,0],".")
        print("The greatest decrease in profits is $", df["Amount_changed"].min(), "on", df.iloc[k,0],".", file=f)
        
f.close()
```

```python
import os
import pandas as pd
import csv

#----------------------------PyPoll----------------------------
df=pd.read_csv("election_data.csv")


#export a text file with the results

path = 'Election Results.txt'
f = open(path, 'w')
print("Election Results")
print("Election Results", file=f)

print("----------------------------")
print("----------------------------", file=f)

# The total number of votes cast
total_ballot=df["Ballot ID"].count()
print("Total Votes:",total_ballot)
print("Total Votes:",total_ballot, file=f)


print("----------------------------")
print("----------------------------", file=f)

Diana_count=df["Candidate"]=="Diana DeGette"
Charles_count=df["Candidate"]=="Charles Casper Stockham"
Raymon_count=df["Candidate"]=="Raymon Anthony Doane"

len(df[Diana_count])
len(df[Charles_count])
len(df[Raymon_count])

print("Diana DeGette :", round(len(df[Diana_count])/total_ballot*100,3), "%", "(", len(df[Diana_count]),")")
print("Charles Casper Stockham :", round(len(df[Charles_count])/total_ballot*100,3), "%", "(", len(df[Charles_count]),")")
print("Raymon Anthony Doane :", round(len(df[Raymon_count])/total_ballot*100,3), "%", "(", len(df[Raymon_count]),")")

print("Diana DeGette :", round(len(df[Diana_count])/total_ballot*100,3), "%", "(", len(df[Diana_count]),")", file=f)
print("Charles Casper Stockham :", round(len(df[Charles_count])/total_ballot*100,3), "%", "(", len(df[Charles_count]),")", file=f)
print("Raymon Anthony Doane :", round(len(df[Raymon_count])/total_ballot*100,3), "%", "(", len(df[Raymon_count]),")", file=f)

print("----------------------------")
print("----------------------------", file=f)

print("Winner: Diana DeGette")
print("Winner: Diana DeGette", file=f)

print("----------------------------")
print("----------------------------", file=f)
```
